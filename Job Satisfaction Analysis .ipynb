```python
def update_vals(element):
    if pd.isnull(element):
        return np.nan
    elif element == "-":
        return False
    else:
        return True
```

## Now that we have two data frames with the same columns, we can combine them with concat. We will differentiate the two columns by the institute column, which we will define next

```python
tafe_resignations_updated['institute'] = 'TAFE'
dete_resignations['institute'] = 'DETE'
```

```python
combined = pd.concat([tafe_resignations_updated,dete_resignations],axis=0)
combined.head()
```

```python
null_counts = combined.isnull().sum().sort_values(ascending = False)
null_counts 
```

```python
null_counts = dict(null_counts)
null_columns_to_delete = []
for key in null_counts: 
    if null_counts[key] > 500:
        null_columns_to_delete.append(key)
combined.drop(columns=null_columns_to_delete, inplace=True)
```

```python
combined.isnull().sum().sort_values(ascending = False)
```

```python
combined["institute_service"] = combined["institute_service"].astype(str)
combined["institute_service"].value_counts(dropna=False).sort_index(ascending=False)
```

```python
def new_updates(element):
    if element == 'More than 20 years':
        return "20"
    elif element == 'Less than 1 year':
        return "1"
    elif "-" in element:
        return element.split("-")[-1] 
    else:
        return element
```

```python
combined["institute_service"] = combined["institute_service"].apply(new_updates)
```

```python
combined["institute_service"].value_counts().sort_index()
```

```python
combined["institute_service"] = combined["institute_service"].astype(float)
combined["institute_service"].head()
```

```python
def categories(element):
    if element < 3:
        return "New: Less than 3 years at a company"
    elif  3 <= element <= 6:
        return "Experienced: 3-6 years at a company"
    elif 3 <= element <= 6:
        return "Established: 7-10 years at a company"
    elif element > 11:
        return "Veteran: 11 or more years at a company"
    else:
        return "Service Dates Not Listed"
        
```

```python
combined["service_cat"] = combined["institute_service"].apply(categories)
```

```python
combined[["service_cat","institute_service"]].head()
```

```python
combined["dissatisfied"].value_counts(dropna=False)
```

```python
combined["dissatisfied"] = combined["dissatisfied"].fillna(False)
combined["dissatisfied"].value_counts(dropna=False)
```

```python
combined_pivot = combined.pivot_table(index="service_cat", values="dissatisfied")
combined_pivot
```

```python
combined_pivot.plot(kind='barh',legend=False,color="b")
```

## Veteran" with 11 or more years at a company, as indicated by the highest probability value of 0.484848. On the other hand, individuals categorized as "New" with less than 3 years at a company have the lowest probability at 0.295337. The group with service dates not listed falls in between, with a probability of 0.389610. Overall, this suggests a trend where the level of experience at a company is positively correlated with the likelihood of the specified outcome.
